```python
import pandas as pd
import os  # Esta librería nos permite trabajar con carpetas

# 1. Definir la URL y la ruta de destino
url = "https://storage.googleapis.com/breathecode/project-files/bank-marketing-campaign-data.csv"
target_dir = "data/raw"  # Solo la carpeta
target_file = "bank_marketing_campaign_data.csv" # Solo el nombre del archivo
target_path = os.path.join(target_dir, target_file) # Ruta completa y segura


df = pd.read_csv(url, sep=';')

# Ahora sí, probamos de nuevo ver la variable objetivo
print("Distribución de la variable objetivo (y):")
print(df['y'].value_counts(normalize=True) * 100)

# 2. Cargar los datos desde la nube
print("Cargando datos desde la URL...")
df = pd.read_csv(url, sep=';')
print("Datos cargados exitosamente.")

# 3. CREAR LA CARPETA SI NO EXISTE
if not os.path.exists(target_dir):
    print(f"La carpeta '{target_dir}' no existe. Creándola...")
    os.makedirs(target_dir) #makedirs crea todas las carpetas intermedias (data, raw)
    print("Carpeta creada.")
else:
    print(f"La carpeta '{target_dir}' ya existe.")

# 4. Guardar una copia en la carpeta 'raw'
print(f"Guardando archivo en '{target_path}'...")
df.to_csv(target_path, index=False)

print(f"¡Hecho! Archivo guardado exitosamente en: {target_path}")

# Opcional: ver las primeras filas para confirmar que todo está bien
print("\nPrimeras filas del dataset cargado:")
display(df.head())






```

```python
# 2 Realizar el EDA "Objetivo predecir la variable y (en donde el cliente contrata el depósito o no)"

# 1. Dimensiones del dataset
print(f"El dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas.")

# 2. Verificar tipos de datos y valores nulos
print("\nInformación del dataset:")
print(df.info())

# 3. Resumen estadístico de variables numéricas (age, duration, campaign, etc.)
print("\nResumen estadístico:")
display(df.describe())

# 4. Ver cómo está repartida nuestra meta (la variable 'y')
print("Distribución de la variable objetivo (y):")
print(df['y'].value_counts(normalize=True) * 100)
```

```python
# EDA 2.1 Limpieza de datos: Eliminar valores nulos y duplicados 

# Revisar nulos
print("Valores nulos por columna:")
print(df.isnull().sum())

# Revisar si hay filas duplicadas
print(f"\nFilas duplicadas encontradas: {df.duplicated().sum()}")
```

```python
# 1. ELIMINAR DUPLICADOS
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
final_rows = df.shape[0]

print(f"Filas iniciales: {initial_rows}")
print(f"Filas eliminadas (duplicados): {initial_rows - final_rows}")
print(f"Filas actuales: {final_rows}")

# 2. TRATAR VALORES NULOS
# Cuántos hay por columna
null_counts = df.isnull().sum()
print("\nConteo de valores nulos por columna:")
print(null_counts[null_counts > 0] if null_counts.sum() > 0 else "No se encontraron valores nulos.")

# ESTRATEGIA DE LIMPIEZA (Solo si existen nulos):
# Para variables numéricas: llenamos con la mediana
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Para variables categóricas (objetos): llenamos con la moda (el valor más frecuente)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\n¡Limpieza completada! Todos los nulos han sido tratados.")
```

```python
# EDA 2.2 Convertir variables categoricas en numericas 

# 1. Separamos la variable objetivo 'y' para tratarla distinto
# La convertimos en 1 (si) y 0 (no)
df['y'] = df['y'].map({'yes': 1, 'no': 0})

# 2. Identificamos las columnas categóricas (las que son tipo 'object')
# Excluimos 'y' porque ya la convertimos a número arriba
categorical_cols = df.select_dtypes(include=['object']).columns

# 3. Aplicamos One-Hot Encoding con get_dummies
# drop_first=True elimina una columna de cada categoría para evitar redundancia
df_final = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# 5. Verificamos el resultado
print(f"Dimensiones originales: {df.shape}")
print(f"Nuevas dimensiones tras get_dummies: {df_final.shape}")
print("\nPrimeras 5 filas del nuevo DataFrame:")
display(df_final.head())
```

```python
# 1. Convertimos a flotantes primero (esto sí acepta NaN) y luego a enteros
# Si hay un NaN, se mantendrá como tal sin romper el código
df_final = df_final.apply(pd.to_numeric, errors='coerce')

# 2. Por seguridad, si quedaron nulos tras el get_dummies, los llenamos con 0
df_final = df_final.fillna(0).astype(int)

# 3. Verificamos el resultado
display(df_final.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 1. Separamos las características (X) de la etiqueta objetivo (y)
X = df_final.drop(columns=['y'])
y = df_final['y']

# 2. Inicializamos el escalador
scaler = StandardScaler()

# 3. Ajustamos y transformamos las características
# Esto convertirá todos los números a una escala donde la media es 0 y la desviación es 1
X_scaled = scaler.fit_transform(X)

# 4. Convertimos X_scaled de vuelta a un DataFrame 
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("¡Datos escalados con éxito!")
print("\nAsí se ven las primeras filas escaladas:")
display(X_scaled.head())
```

```python
# Paso 3 Modelo de Regresión Logistica

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# 1. Volvemos a dividir pero usando STRATIFY
# Esto asegura que haya ejemplos de '1' y '0' en ambos grupos
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, 
    test_size=0.2, 
    random_state=42, 
    stratify=y  # <--- ESTA ES LA CLAVE
)

# 2. Creamos el modelo
modelo_balanceado = LogisticRegression(class_weight='balanced', max_iter=1000)
modelo_balanceado.fit(X_train, y_train)

# 3. Entrenamos
modelo_balanceado.fit(X_train, y_train)

print("¡Ahora sí! Modelo entrenado exitosamente con ambas clases.")

# 4. Predicciones y Reporte
y_pred = modelo_balanceado.predict(X_test)
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# 1. Calcular la matriz de confusión con los datos reales y las predicciones

cm = confusion_matrix(y_test, y_pred)

# 2. Configurar el tamaño del gráfico
plt.figure(figsize=(8, 6))

# 3. Crear el heatmap con Seaborn
# labels=[0, 1] asegura que el orden sea correcto
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['No Contrata (0)', 'SÍ Contrata (1)'],
            yticklabels=['No Contrata (0)', 'SÍ Contrata (1)'])

# 4. Añadir etiquetas y título
plt.xlabel('Predicción del Modelo')
plt.ylabel('Valor Real (Actual)')
plt.title('Matriz de Confusión: Modelo de Regresión Logística')

# 5. Mostrar el gráfico
plt.show()
```

```python
# 5 mejoramiento del modelo

# Entrenamos un nuevo modelo dándole más importancia a la clase '1'
modelo_bal = LogisticRegression(class_weight='balanced', max_iter=1000)
modelo_bal.fit(X_train, y_train)

# Nuevas predicciones
y_pred_bal = modelo_bal.predict(X_test)

# Mira la diferencia en el reporte
from sklearn.metrics import classification_report
print("Reporte del Modelo Balanceado:")
print(classification_report(y_test, y_pred_bal))

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Creamos un DataFrame con los nombres de las columnas y sus coeficientes
# Usamos X_scaled.columns porque son los nombres de las variables que el modelo vió
coef_df = pd.DataFrame({
    'Variable': X_scaled.columns,
    'Coeficiente': modelo_balanceado.coef_[0]
})

# 2. Calculamos el valor absoluto para identificar cuáles influyen más (sin importar el signo)
coef_df['Abs_Coef'] = coef_df['Coeficiente'].abs()
coef_df = coef_df.sort_values(by='Abs_Coef', ascending=False)

# 3. Graficamos las 15 variables más importantes
plt.figure(figsize=(10, 8))
# Usamos el coeficiente real (con signo) para ver si ayuda o perjudica
sns.barplot(data=coef_df.head(15), x='Coeficiente', y='Variable', palette='coolwarm')

plt.title('Top 15 Variables que más influyen en la predicción')
plt.axvline(0, color='black', linestyle='-', linewidth=1) # Línea en el cero para separar pos/neg
plt.xlabel('Peso del Coeficiente (Impacto)')
plt.ylabel('Variables del Dataset')
plt.grid(axis='x', alpha=0.3)
plt.show()

# 4. Mostramos el top 10 numérico
print("Top 10 variables con mayor peso:")
display(coef_df[['Variable', 'Coeficiente']].head(10))
```

```python
# Interpretación de los resultados:
# duration (1.80) es la variable que más influye positivamente en que un cliente contrate el depósito, es decir, a mayor duración de la llamada, más probable es que contrate.
# month_may (-0.64) muestra que las llamadas realizadas en mayo tienen un impacto negativo en la contratación, lo que podría indicar que las campañas en ese mes no son tan efectivas o que los clientes contactados en mayo tienen menos interés.
# euribor3m (0.71) es la tasa de interés a 3 meses, y su coeficiente positivo indica que a mayor euribor3m, más probable es que el cliente contrate el depósito, lo que tiene sentido ya que un euribor más alto puede hacer que los depósitos sean más atractivos.
# nr_employed (-1.26) es el número de empleados, y su coeficiente negativo indica que a mayor número de empleados, menos probable es que el cliente contrate el depósito, lo que podría sugerir que clientes con trabajos más estables o en empresas más grandes tienen menos interés en contratar este tipo de producto.
```

```python
# 5.1 Optimizar el parametro C 

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# 1. Aseguramos que los datos estén divididos (por si se borraron de memoria)
# Usamos X_scaled y y que definiste al principio
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, 
    test_size=0.2, 
    random_state=42, 
    stratify=y
)

# 2. Definir los parámetros
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l2']
}

# 3. Inicializar el modelo base
logreg = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)

# 4. Configurar y ejecutar la búsqueda
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='f1', verbose=1)

print("Iniciando búsqueda del mejor parámetro C...")
grid_search.fit(X_train, y_train)

# 5. Mostrar resultados y crear el mejor modelo
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)

print(f"\n¡Mejor parámetro C encontrado: {grid_search.best_params_['C']}")
print("\nReporte de Clasificación del Modelo Optimizado:")
print(classification_report(y_test, y_pred_best))

```

```python
# Análisis de los resultados tras optimizar el parámetro C:
# 1. El salto en el Recall  para la clase 1 (los que sí contratan) era de 0.43.Ahora: ¡Subió a 0.90! Qué significa: Antes el modelo perdía aun % importante de los clientes potenciales. Ahora, el modelo detecta al 90% de las personas que realmente quieren el depósito. 
# 2. El equilibrio del F1-Score para la clase 1 subió a 0.60.Encontrar el mejor parámetro $C = 0.1$ ayudó a que el modelo no fuera ni muy simple ni muy complejo, logrando ese balance donde capturamos a casi todos los clientes interesados.
# 3. El "Costo" la precisión bajó a 0.44. Esto quiere decir que de cada 10 llamadas que el modelo sugiere, unas 4 o 5 terminarán en contrato. 
```

```python
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

def obtener_metricas(y_true, y_pred, nombre_modelo):
    # Obtenemos precisión, recall y f1 para la clase 1 (los que sí contratan)
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='binary')
    acc = accuracy_score(y_true, y_pred)
    
    return {
        'Modelo': nombre_modelo,
        'Accuracy (Global)': f"{acc:.2%}",
        'Precision (Clase 1)': f"{precision:.2%}",
        'Recall (Clase 1)': f"{recall:.2%}",
        'F1-Score (Clase 1)': f"{f1:.2%}"
    }

# 1. Calculamos métricas para el modelo optimizado
metricas_opt = obtener_metricas(y_test, y_pred_best, "Modelo Optimizado (GridSearch)")

# 2. Intentamos recuperar las métricas del modelo inicial 
# (Usamos los valores que vimos en tus capturas anteriores para comparar)
metricas_ini = {
    'Modelo': 'Modelo Inicial (Base)',
    'Accuracy (Global)': '91.00%',
    'Precision (Clase 1)': '66.00%',
    'Recall (Clase 1)': '43.00%',
    'F1-Score (Clase 1)': '52.00%'
}

# 3. Creamos la tabla comparativa
df_comparacion = pd.DataFrame([metricas_ini, metricas_opt])

print("Comparativa de Rendimiento:")
display(df_comparacion)
```
